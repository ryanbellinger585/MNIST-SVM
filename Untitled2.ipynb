function to aid in the selecting of training batches.

```python
def picker(size, label, labels):
  toReturn = [0]*size
  i = 0
  while i < size:
    tryto = randrange(start = 0,stop = 60000)
    if tryto not in toReturn and labels[tryto] != label:
      toReturn[i] = tryto
      i+=1
  return toReturn
```

Creates k folds

```python
def folder(bigBatch, folds):
  toReturn1 = []
  toReturn2 = []
  np.random.shuffle(bigBatch)
  n = bigBatch.shape[0]//folds
  for i in range(folds):
    toReturn1.append(bigBatch[n*i:n*(i+1)])
    toReturn2.append(np.append(bigBatch[:n*i], bigBatch[n*(i+1):], axis = 0))
  return toReturn1, toReturn2
```

The SVM. Learning rate .01. 1 norm soft margin, linear kernel.

```python
def wbTrainer(batch, ys, c, learn):
  l = batch.shape[0]
  w = np.zeros(batch.shape[1])
  finalw = w
  b = 0
  finalb = b
  radius = 0
  lagrangian = maxsize
  errorCounter = 0
  k = 0
  for i in batch:
    radius = max(sqrt(np.dot(i,i)),radius)
  while (errorCounter <= 216) or k == 0:
    for j in range(l):
      sum = 0
      if ys[j]*(np.dot(w, batch[j])+b) <= 0:
        w = np.add(w, learn*ys[j]*batch[j])
        b += learn*ys[j]*radius**2
        k+=1
    errorSum = 0
    maxMargin = 1
    for n in range(l):
      errorSum += max(0,maxMargin - ys[n]*(np.dot(batch[n],w)+b))
    newLagrangian = abs(c*errorSum+np.dot(w,w))
    if(newLagrangian >= lagrangian):
      errorCounter += 1
    else:
      lagrangian = newLagrangian
      errorCounter = 0
      finalw = w
      print('check')
      finalb = b
  print(finalb, finalw)
  return finalw, finalb
```

Given in the problem description:

$\left<w\cdot w \right> + C\sum_{i=1}^{l}\xi_{i}$ subject to $(y_i\left<w\cdot x_i\right>+b) \geq 1-\xi _i$

The Lagrangian form of the problem:

$L = \left<w\cdot w \right> + C\sum_{i=1}^{l}\xi_{i} + \sum_{j=1}^{l}\alpha[1-xi_{i}-(y_i\left<w\cdot x_i\right>+b)]$

Taking the partial derivative with respect to w to find weight in dual form

$\frac{\partial L}{\partial w} = 2w-\sum_{i=1}^{l}\alpha _iy_ix_i = 0$

$w = \frac{\sum_{i=1}^{l}\alpha _iy_ix_i}{2}$

Taking the partial derivative with respect to $\xi\$

$\frac{\partial L}{\partial \xi} = C + \sum_{i=1}^l\alpha_i = 0$

Taking the partial derivative with respect to b

$\frac{\partial L}{\partial b} = \sum_{i=1}^l\alpha_i = 0$

Substituting into found values into primal problem to find dual problem 

$W = \frac{\sum_{i,j = 1}^{l}\alpha _i\alpha _jy_iy_j\left<x_i\cdot x_j \right>}{4}-\frac{\sum_{i,j = 1}^{l}\alpha _i\alpha _jy_iy_j\left<x_i\cdot x_j \right>}{2}+\sum_{i = 1}^l\alpha_i$

$W = \sum_{i = 1}^l\alpha_i-\frac{\sum_{i,j = 1}^{l}\alpha _i\alpha _jy_iy_j\left<x_i\cdot x_j \right>}{4}$

In the primal problem the problem is represented by 1 and in the dual problem it is represented by $\sum_{i = 1}^l\alpha_i$. The margin is helpful in making sure fringe cases aren't misclassified and for making sure slight variations from the classes don't cause misclassification. The support vectors are represented by $\alpha_iy_ix_i$ and are used to train the model. When you solve the dual problem instead of the primal problem it makes implementing a kernel function much more straightforward than when solving the primal function, as the dot product in the dual function can simply be replaced by whatever kernel function you choose.
